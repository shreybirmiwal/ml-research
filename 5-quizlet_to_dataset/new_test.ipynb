```python
import time
import random
import jsonlines
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException

def click_button_if_exists(driver):
    try:
        # Wait for the button to be clickable
        button = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CLASS_NAME, 'AssemblyButtonBase')))
        
        # Scroll to the button's location
        driver.execute_script("arguments[0].scrollIntoView();", button)
        
        # Click the button
        button.click()
        print("Button clicked successfully.")
    except NoSuchElementException:
        print("Button not found. Skipping click.")
    except Exception as e:
        print(f"Error clicking button: {str(e)}")

def scrape_quizlet(file_path, output_file):
    chrome_options = Options()
    driver = webdriver.Chrome(options=chrome_options)

    try:
        driver.get(f"file://{file_path}")
        click_button_if_exists(driver)

        print(driver.page_source)

        terms = []
        definitions = []
        
        term_elements = driver.find_elements(By.CLASS_NAME, 'SetPageTerms-term')
        print(term_elements.text)
        print("*")

        for term_elem in term_elements:
            print(term_elem[0].text)
            print("**")
            term_x = term_elem.find_element(By.XPATH, ".//div[@data-testid='set-page-card-side'][1]/div/span/span")
            term = term_x.text.strip()

            def_x = term_elem.find_element(By.XPATH, ".//div[@data-testid='set-page-card-side'][2]/div/span/span")
            definition = def_x.text.strip()

            terms.append(term)
            definitions.append(definition)
        
        with jsonlines.open(output_file, mode='a') as writer:
            for term, definition in zip(terms, definitions):
                data = {'input': term, 'output': definition}
                writer.write(data)
        
        print(f"Scraping completed for {file_path}. Data appended to {output_file}")
    
    except Exception as e:
        print(f"Error occurred during scraping for {file_path}: {str(e)}")
    
    finally:
        driver.quit()

def scrape_multiple_quizlets(file_paths, output_file):
    for i, path in enumerate(file_paths):
        scrape_quizlet(path, output_file)

# List of paths to local HTML files
file_paths = [
    r'C:\Users\shrey\projects\ml-projects\5-quizlet_to_dataset\page2.html'
]

output_file = 'APES_dataset.jsonl'
scrape_multiple_quizlets(file_paths, output_file)




```
